```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
base_data = '../../output'

set_1_facenet = pd.read_csv(f'{base_data}/set_1/facenet/metrics.csv')
set_1_facenet
```

```python
set_1_mobilenet = pd.read_csv(f'{base_data}/set_1/mobilenet/metrics.csv')
set_1_mobilenet
```

```python

set_2_facenet = pd.read_csv(f'{base_data}/set_2/facenet/metrics.csv')
set_2_facenet
```

```python

set_2_mobilenet = pd.read_csv(f'{base_data}/set_2/mobilenet/metrics.csv')
set_2_mobilenet
```

```python

set_10_facenet = pd.read_csv(f'{base_data}/set_10/facenet/metrics.csv')
set_10_facenet
```

```python

set_10_mobilenet = pd.read_csv(f'{base_data}/set_10/mobilenet/metrics.csv')
set_10_mobilenet
```

```python
fig = plt.figure(figsize=(8, 10))

set_1_plot_mobile = set_1_mobilenet[['set','accuracy detection']].drop_duplicates() 
set_1_plot_facenet = set_1_facenet[['set','accuracy detection']].drop_duplicates()

set_2_plot_mobile = set_2_mobilenet[['set','accuracy detection']].drop_duplicates()
set_2_plot_facenet = set_2_facenet[['set','accuracy detection']].drop_duplicates()

set_10_plot_mobile = set_10_mobilenet[['set','accuracy detection']].drop_duplicates()
set_10_plot_facenet = set_10_facenet[['set','accuracy detection']].drop_duplicates()

data_mobile = set_1_plot_mobile
data_mobile= data_mobile.append(set_2_plot_mobile, ignore_index = True) 
data_mobile =data_mobile.append(set_10_plot_mobile, ignore_index = True) 

data_facenet = set_1_plot_facenet
data_facenet = data_facenet.append(set_2_plot_facenet, ignore_index = True) 
data_facenet = data_facenet.append(set_10_plot_facenet, ignore_index = True) 

ax = data_mobile.plot(x="set", y="accuracy detection", kind="bar",   position=0, width=0.3)
data_facenet.plot(x="set", y="accuracy detection", kind="bar",  ax=ax, color="C2", position=1, width=0.3)

ax.set_xlabel("Localization")
ax.set_ylabel("Acc Detection")
ax.set_title("Acc Detection")
ax.legend(["MobileNet", "Facenet"])
```

```python
set_1_plot_mobile = set_1_mobilenet[['user','location','accuracy recognition']]
set_1_plot_facenet = set_1_facenet[['user','location','accuracy recognition']]

set_1_plot_mobile.loc[:,"period"] = ['-'.join(i) for i in zip(set_1_plot_mobile["location"].map(str),set_1_plot_mobile["user"])]
set_1_plot_facenet.loc[:,"period"] = ['-'.join(i) for i in zip(set_1_plot_facenet["location"].map(str),set_1_plot_facenet["user"])]

fig, axs = plt.subplots(3,figsize=(8, 10))
fig.tight_layout() 
set_1_plot_mobile.plot(x="period", y="accuracy recognition", kind="bar",ax=axs[0],  position=0, width=0.3)
set_1_plot_facenet.plot(x="period", y="accuracy recognition", kind="bar", ax=axs[0], color="C2", position=1, width=0.3)

axs[0].set_xlabel("Localization")
axs[0].set_ylabel("Acc Recognition")
axs[0].set_title("Set 1")
axs[0].legend(["MobileNet", "Facenet"])

plt.subplots_adjust(hspace = .7)

set_2_plot_mobile = set_2_mobilenet[['user','location','accuracy recognition']]
set_2_plot_facenet = set_2_facenet[['user','location','accuracy recognition']]

set_2_plot_mobile.loc[:,"period"] = ['-'.join(i) for i in zip(set_2_plot_mobile["location"].map(str),set_2_plot_mobile["user"])]
set_2_plot_facenet.loc[:,"period"] = ['-'.join(i) for i in zip(set_2_plot_facenet["location"].map(str),set_2_plot_facenet["user"])]

set_2_plot_mobile.plot(x="period", y="accuracy recognition", kind="bar",ax=axs[1],  position=0, width=0.3)
set_2_plot_facenet.plot(x="period", y="accuracy recognition", kind="bar", ax=axs[1], color="C2", position=1, width=0.3)

axs[1].set_xlabel("Localization")
axs[1].set_ylabel("Acc Recognition")
axs[1].set_title("Set 2")
axs[1].legend(["MobileNet", "Facenet"])

plt.subplots_adjust(hspace = .7)

set_10_plot_mobile = set_10_mobilenet[['user','location','accuracy recognition']]
set_10_plot_facenet = set_10_facenet[['user','location','accuracy recognition']]

set_10_plot_mobile.loc[:,"period"] = ['-'.join(i) for i in zip(set_10_plot_mobile["location"].map(str),set_10_plot_mobile["user"])]
set_10_plot_facenet.loc[:,"period"] = ['-'.join(i) for i in zip(set_10_plot_facenet["location"].map(str),set_10_plot_facenet["user"])]

set_10_plot_mobile.plot(x="period", y="accuracy recognition", kind="bar",ax=axs[2],  position=0, width=0.3)
set_10_plot_facenet.plot(x="period", y="accuracy recognition", kind="bar", ax=axs[2], color="C2", position=1, width=0.3)

axs[2].set_xlabel("Localization")
axs[2].set_ylabel("Acc Recognition")
axs[2].set_title("Set 10")
axs[2].legend(["MobileNet", "Facenet"])
```
