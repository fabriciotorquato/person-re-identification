```python
fig = plt.figure(figsize=(8, 10))

set_1_plot_mobile = mobilenet_data[0][['set','accuracy detection']].drop_duplicates() 
set_1_plot_facenet = facenet_data[0][['set','accuracy detection']].drop_duplicates()

set_2_plot_mobile = mobilenet_data[1][['set','accuracy detection']].drop_duplicates()
set_2_plot_facenet = facenet_data[1][['set','accuracy detection']].drop_duplicates()

set_10_plot_mobile = mobilenet_data[2][['set','accuracy detection']].drop_duplicates()
set_10_plot_facenet = facenet_data[2][['set','accuracy detection']].drop_duplicates()

data_mobile = set_1_plot_mobile
data_mobile= data_mobile.append(set_2_plot_mobile, ignore_index = True) 
data_mobile =data_mobile.append(set_10_plot_mobile, ignore_index = True) 

data_facenet = set_1_plot_facenet
data_facenet = data_facenet.append(set_2_plot_facenet, ignore_index = True) 
data_facenet = data_facenet.append(set_10_plot_facenet, ignore_index = True) 

ax = data_mobile.plot(x="set", y="accuracy detection", kind="bar",   position=0, width=0.3)
data_facenet.plot(x="set", y="accuracy detection", kind="bar",  ax=ax, color="C2", position=1, width=0.3)

ax.set_xlabel("Conjunto")
ax.set_ylabel("Acurácia na detecção")
ax.set_title("Acurácia na detecção")
ax.legend(["MobileNet", "Facenet"])
```

```python

def plot_chart(axis_y,ax,idx,ylabel,title):

  set_plot_mobile = mobilenet_data[idx][['user','location',axis_y]]
  set_plot_facenet = facenet_data[idx][['user','location',axis_y]]

  set_plot_mobile.loc[:,"period"] = ['-'.join(i) for i in zip(set_plot_mobile["location"].map(str),set_plot_mobile["user"])]
  set_plot_facenet.loc[:,"period"] = ['-'.join(i) for i in zip(set_plot_facenet["location"].map(str),set_plot_facenet["user"])]

  set_plot_mobile = set_plot_mobile.drop(['user','location'],axis=1)
  set_plot_facenet = set_plot_facenet.drop(['user','location'],axis=1)

  set_plot_mobile.index = set_plot_mobile["period"]
  set_plot_facenet.index = set_plot_facenet["period"]

  x = list(set(set_plot_facenet[set_plot_facenet[axis_y]>0]["period"].append(set_plot_mobile[set_plot_mobile[axis_y]>0]['period']).values))
  set_plot_mobile = set_plot_mobile.reindex(x, fill_value=0)
  set_plot_facenet = set_plot_facenet.reindex(x, fill_value=0)

  set_plot_mobile.plot(y=axis_y, kind="bar", ax=ax ,position=0, width=0.3)
  set_plot_facenet.plot(y=axis_y, kind="bar",  ax=ax , color="C2", position=1, width=0.3)

  ax.set_xlabel("Localização")
  ax.set_ylabel(ylabel)
  ax.set_title(title)
  ax.legend(["MobileNet", "Facenet"])

  labels = [item.get_text().replace("space","Câmera") for item in ax.get_xticklabels()]
  ax.set_xticklabels(labels)

  return ax
```

```python
fig, axs = plt.subplots(3,figsize=(8, 10))
fig.tight_layout() 

axs[0] = plot_chart('accuracy recognition',axs[0],0,"Acurácia em reconhecimento facial","Conjunto 1")
plt.subplots_adjust(hspace = .8)

axs[1] = plot_chart('accuracy recognition',axs[1],1,"Acurácia em reconhecimento facial","Conjunto 2")
plt.subplots_adjust(hspace = .8)

axs[2] = plot_chart('accuracy recognition',axs[2],2,"Acurácia em reconhecimento facial","Conjunto 10")
plt.subplots_adjust(hspace = .8)

plt.plot()
```
